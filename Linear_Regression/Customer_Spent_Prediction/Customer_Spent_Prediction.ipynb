```python
 from sklearn.model_selection import train_test_split
```

```python
X = df[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
y = df['Yearly Amount Spent']
```

```python
X,y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state =42)
```

```python
X_train
```

```python
## training model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.coef_
```

```python
cdf = pd.DataFrame(lm.coef_,X.columns,columns=['Coef'])
```

```python
print(cdf)
```

```python
## the coeffiecient with the higher value is the coefficient of the feature having greater impact on the target variable
```

```python
## predictions
```

```python
predictions = lm.predict(X_test)
```

```python
print(predictions)
```

```python
sns.scatterplot(x=predictions,y= y_test)
plt.xlabel("Predictions")
plt.title("Evaluation of our LM model")
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
import math
```

```python
print(mean_squared_error(y_test,predictions))
```

```python
print(mean_absolute_error(y_test,predictions))
```

```python
print(math.sqrt(mean_squared_error(y_test,predictions)))
```

```python
## Residuals
```

```python
Residuals = y_test - predictions
```

```python
sns.displot(Residuals,bins=30,kde=True)
```

```python
##normality
import pylab
import scipy.stats as stats

stats.probplot(Residuals,dist='norm',plot=pylab)
pylab.show()
```
